**Step 2:** Develop the IP2AS tool - This tool maps an IP address to an AS. It uses static table address prefix to AS number collected from whois DB and BGP tables. This table is stored in a file and should be given to the tool as a parameter. It will perform longest prefix matching and will map the IP to an AS number. The tool should print out the longest prefix that the IP address is matched to, and the corresponding AS number.

 
Steps:

1. Put the set of IP addresses you want to map to ASes into the `<IP
   file>`. You can list one IP address per line. 
   
   For example, look at IPlist.txt file, which contains the following:
   
```
   169.237.33.90
   208.30.172.70
```

2. The `<DB file>` has data about which address block belongs to
   a particular AS (look at `DB_091803_v1.txt` file, for example).
   The `<DB file>` is constructed based on IRR database and
   BGP routing table.
  
3. Run ip2as and specify the <DB file> and <IP file>

   For example, if you run your code, the output should look like the following:
```   
   169.237.0.0/16  1852 169.237.33.90
   208.0.0.0/11  1239 208.30.172.70
```

````python
!pip install IPy
# import IPy to eazliy manage IP information
import sys
from IPy import IP

# Get data from IPlist
Iplist = open("IPlist.txt", 'r')
input = Iplist.read().splitlines()
Iplist.close()

# Get data from DB file
DB = open("DB_091803_v1.txt", 'r')
DB_match = DB.read().splitlines()
DB.close()
````

````python
# The code should assume that the variables db_filename and input_filename 
# contain the filenames of the database and inputs respectively. 
#
# Model output is provided for you.
#
# Your code with documentation


class Data:
    def __init__(self,_prefixlen,_ipSub,_data):
        self.prefixlen = _prefixlen
        self.ipSub = _ipSub
        self.data = _data
# Implement IPy function

ipMap: dict = {}
# dict for store ip imformation

for line in DB_match:
    spliteds = line.split(' ')

    try:
        ipSub = IP(spliteds[0] + "/" + spliteds[1])
    except ValueError:
        pass
        # If value is invalid, pass to next loop
    else:
        for ip in input:
            if ip in ipSub:
                if (not ipMap.__contains__(ip)) or ipSub.prefixlen() >= ipMap[ip].prefixlen:
                  # Match the prefix and store it into dict, store the larger metric ip
                    ipMap[ip] = Data(ipSub.prefixlen(), ipSub, spliteds[2])

for ip,d in ipMap.items():
    print("{} {} {}".format(d.ipSub, d.data, ip))
````
